```python
import torch
import json
from one_peace.models import from_pretrained

id2label = json.load(open("assets/vggsound_id2label.json"))

device = "cuda" if torch.cuda.is_available() else "cpu"
model = from_pretrained(
  "ONE-PEACE_VGGSound",
    model_type="one_peace_classify",
    device=device,
    dtype="float32"
)

# process audio
audio_list = ["assets/cow.flac", "assets/dog.flac", "assets/1_ambiente_inicio.wav"]
src_audios, audio_padding_masks = model.process_audio(audio_list)

with torch.no_grad():
    # extract audio features
    audio_logits = model.extract_audio_features(src_audios, audio_padding_masks)
    print(audio_logits.size())
    predict_label_ids = audio_logits.argmax(1).cpu().tolist()

for audio, predict_label_id in zip(audio_list, predict_label_ids):
    predict_label = id2label[str(predict_label_id)]
    print('audio: {}, predict label: {}'.format(audio, predict_label))
```

```python
import torch
import json
from one_peace.models import from_pretrained

device = "cuda" if torch.cuda.is_available() else "cpu"
model_path="/content/drive/MyDrive/2024dcase/task8_retrieval/pretrained_chkpoint/finetune_al_retrieval_onepiece.pt"
model = from_pretrained(
    # "ONE-PEACE_VGGSound",  # Use the AudioCaps pretrained model for text-audio retrieval
    model_path,
    model_type="one_peace_retrieval",
    device=device,
    dtype="float32"
)

# Process text queries
text_queries = ["A dog barking", "A cow mooing", "Background noise"]
src_texts, text_padding_masks = model.process_text(text_queries)

# Process audio files
audio_list = ["assets/dog.flac", "assets/cow.flac", "assets/1_ambiente_inicio.wav"]
src_audios, audio_padding_masks = model.process_audio(audio_list)

with torch.no_grad():
    # Extract text features
    text_logits = model.extract_text_features(src_texts, text_padding_masks)

    # Extract audio features
    audio_logits = model.extract_audio_features(src_audios, audio_padding_masks)

    # Compute similarity scores between text and audio features
    similarity_scores = model.compute_similarity(text_logits, audio_logits)

# Print the top-ranked audio for each text query
for i, text_query in enumerate(text_queries):
    sorted_indices = torch.argsort(similarity_scores[i], descending=True)
    print(f"Text Query: {text_query}")
    for j in sorted_indices:
        print(f"  {audio_list[j]}: {similarity_scores[i][j]:.2f}")
```
