```python
import os

# Check if the file paths in the 'file_name' column exist and if they have read permissions
def check_file_paths(df, column_name):
    results = []
    for file_path in df[column_name]:
        exists = os.path.exists(file_path)
        readable = os.access(file_path, os.R_OK) if exists else False
        results.append((file_path, exists, readable))
    return results

# Perform the check on the new dataset
file_path_results = check_file_paths(df, 'file_name')

# Convert results to a DataFrame for better readability
file_path_df = pd.DataFrame(file_path_results, columns=['file_name', 'exists', 'readable'])

file_path_df.head()
```

```python
import re
def process_text(text, max_length=None):
    text = text.lower().lstrip(",.!?*#:;~")
    text = re.sub(
        r"\s{2,}|\t",
        ' ',
        text,
    )
    text = text.rstrip('\n')
    text = text.strip(' ')

    if max_length is not None:
        text = ' '.join(text.split(' ')[:max_length])

    return text
```

```python
def combine_columns(row, columns):
    # Initialize an empty string to hold the combined result
    result = ""
    # Iterate over each column name provided in the list
    for column in columns:
        # Check if the column value is not NaN
        if not pd.isna(row[column]):
            # Append the column value to the result string, prefixed by a space if result is not empty
            result += (" " if result else "") + str(row[column])
    return result.strip()
```

```python
columns_to_combine = ['caption_1', 'caption_2', 'caption_3', 'caption_4', 'caption_5', 'keywords']
df_clo['caption'] = df_clo.apply(combine_columns, columns=columns_to_combine, axis=1)
```

```python
def combine_2rd_columns(col_1, col_2):
    result = col_1
    if not pd.isna(col_2):
        result += " " + str(col_2)
    return result

df_wav["caption"] = df_wav.apply(lambda x: combine_2rd_columns(x['description'], x['tags']), axis=1)
# df_wav["caption"] = df_wav.apply(lambda x: combine_2rd_columns(x['caption'], x['title']), axis=1)
```
